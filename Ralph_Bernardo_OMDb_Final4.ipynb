```python
import requests
import json
import config
import pandas as pd

OMDb_API_KEY = config.omdb_api_key

# Using the OMDb API available at: https://www.omdbapi.com
# Make a request using two API parameters
movie = 'Superman'
year = 1978

ENDPOINT = "http://www.omdbapi.com/"

PARAMETERS = {'apikey': OMDb_API_KEY,
              't': movie,
              'y': year
             }

response = requests.get(url = ENDPOINT, params = PARAMETERS)
print('Response Code:', response, '\n')
#print ('Data Type:', type(response), '\n')
print('Sample JSON response:','\n')
api_call = response.json()
print(api_call, '\n')
```
