```python
import requests
import json
import config
import pandas as pd

OMDb_API_KEY = config.omdb_api_key

# Using the OMDb API available at: https://www.omdbapi.com
# Make a request using two API parameters
movie = 'Superman'
year = 1978

ENDPOINT = "http://www.omdbapi.com/"

PARAMETERS = {'apikey': OMDb_API_KEY,
              't': movie,
              'y': year
             }

response = requests.get(url = ENDPOINT, params = PARAMETERS)
print('Response Code:', response, '\n')
#print ('Data Type:', type(response), '\n')
print('Sample JSON response:','\n')
api_call = response.json()
print(api_call, '\n')
```

```python
# Request list of DC Extended Cinematic Universe, Elseworlds, and DC Universe movies

movie_titles_and_years = [('Superman', 1978),
                          ('Superman II', 1980),
                          ('Superman III', 1983),
                          ('Superman IV: The Quest for Peace', 1987),
                          ('Batman', 1989),
                          ('Batman Returns', 1992),
                          ('Batman Forever', 1995),
                          ('Batman & Robin', 1997),
                          ('Catwoman', 2004),
                          ('Constantine', 2005),
                          ('Batman Begins', 2005),
                          ('Superman Returns', 2006),
                          ('The Dark Knight', 2008),
                          ('Watchmen', 2009),
                          ('Green Lantern', 2011),
                          ('The Dark Knight Rises', 2012),
                          ('Man of Steel', 2013),
                          ('Batman v Superman: Dawn of Justice', 2016),
                          ('Suicide Squad', 2016),
                          ('Wonder Woman', 2017),
                          ('Justice League', 2017),
                          ('Aquaman', 2018),
                          ('Shazam!', 2019),
                          ('Joker', 2019),
                          ('Birds of Prey (and the Fantabulous Emancipation of One Harley Quinn', 2020), 
                          ('Wonder Woman 1984', 2020),
                          ('The Suicide Squad', 2021),
                          ('The Batman', 2022),
                          ('Black Adam', 2022),
                          ('Shazam! Fury of the Gods', 2023),
                          ('The Flash', 2023),
                          ('Blue Beetle', 2023),
                          ('Aquaman and the Lost Kingdom', 2023),
                          ('Joker: Folie à Deux', 2024),
                          ('Superman', 2025)]

print('We are retrieving', len(movie_titles_and_years), 'DC movies.\n')

movies = []
for title, year in movie_titles_and_years:
    ENDPOINT = f"http://www.omdbapi.com/"

    PARAMETERS = {'apikey': OMDb_API_KEY,
              't': title,
              'y': year
             }
    response = requests.get(url = ENDPOINT, params = PARAMETERS)  
    #print('Response Code:', response, '\n')
    api_call = response.json()
    #print ('Data Type:', type(api_call), '\n')
    #print(api_call)
    #print()
    movies.append(api_call)
```

```python
#Check that all API calls were inserted into a list.

print ('Data Type:', type(movies), '\n')

#print(movies)
```

```python
# Insert DC movies into a DataFrame

dc_movies_display = pd.DataFrame(movies)
dc_movies_display.T
```

```python
# Remove columns to remove what we will analyze

dc_movies_display_filtered = dc_movies_display.drop(columns=['Rated',
                                                             'Released',
                                                             'Runtime',
                                                             'Genre',
                                                             'Director',
                                                             'Writer',
                                                             'Actors',
                                                             'Plot',
                                                             'Language',
                                                             'Country',
                                                             'Awards',
                                                             'Poster',
                                                             'Ratings',
                                                             'Metascore',
                                                             'imdbRating',
                                                             'imdbVotes',
                                                             'imdbID',
                                                             'Type',
                                                             'DVD',
                                                             'BoxOffice',
                                                             'Production',
                                                             'Website',
                                                             'Response'])
```

```python
dc_movies_display_filtered['Year'] = dc_movies_display_filtered['Year'].astype(int)
#dc_movies_display_filtered['Year']
```

```python
dc_movies_box_office = []
for i in range(len(movies)):
     try:
         dc_box_office = int(movies[i]['BoxOffice'].strip('$').replace(',', ''))
         #print(dc_box_office)
     except:
         dc_box_office = 'N/A'
         print('Box Offce unavailable for:', movies[i]['Title'])
     dc_movies_box_office.append(dc_box_office)
#print(dc_movies_box_office)
```

```python
dc_movies_display_filtered['Box Office'] = dc_movies_box_office
#dc_movies_display_filtered['Box Office']
```

```python
# Create list of Rotten Tomatoes scores for DC movies

dc_movies_ratings = []
for i in range(len(movies)):
     try:
         dc_movies_rating = int(movies[i]['Ratings'][1]['Value'].strip('%')) # Rotten Tomatoes score is the 2nd value of the array
     except:
         #dc_movies_rating = 'N/A'
         #print('Rotten Tomatoes Score unavailable for:', movies[i]['Title'])
         if movies[i]['Title'] == 'The Suicide Squad': # 'The Suicide Squad' is missing a Rotten Tomatoes score on the API
             dc_movies_rating = 90
     dc_movies_ratings.append(dc_movies_rating)
#print(dc_movies_ratings)
```

```python
dc_movies_display_filtered['Rotten Tomatoes'] = dc_movies_ratings
#dc_movies_display_filtered['Rotten Tomatoes']
```

```python
# Request list of Marvel Cinematic Universe movies
# Note each Marvel movie has a unique title so the year parameter is not needed

movie_titles = ['Iron Man', 
                'The Incredible Hulk', 
                'Iron Man 2', 
                'Thor', 
                'Captain America: The First Avenger', 
                'The Avengers', 
                'Iron Man 3', 
                'Thor: The Dark World', 
                'Captain America: The Winter Soldier', 
                'Guardians of the Galaxy', 
                'Avengers: Age of Ultron', 
                'Ant-Man',
                'Captain America: Civil War',
                'Doctor Strange',
                'Guardians of the Galaxy Vol. 2',
                'Spider-Man: Homecoming',
                'Thor: Ragnarok',
                'Black Panther',
                'Avengers: Infinity War',
                'Ant-Man and the Wasp',
                'Captain Marvel',
                'Avengers: Endgame',
                'Spider-Man: Far From Home',
                'Black Widow',
                'Shang-Chi and the Legend of the Ten Rings',
                'Eternals',
                'Spider-Man: No Way Home',
                'Doctor Strange in the Multiverse of Madness',
                'Thor: Love and Thunder',
                'Black Panther: Wakanda Forever',
                'Ant-Man and the Wasp: Quantumania',
                'Guardians of the Galaxy Vol. 3',
                'The Marvels',
                'Deadpool & Wolverine',
                'Captain America: Brave New World',
                'Thunderbolts*',
                'Fantastic Four: First Steps']

print('We are retrieving', len(movie_titles), 'Marvel movies.\n')
movies = []
for movie in movie_titles:
    ENDPOINT = f"http://www.omdbapi.com/"

    PARAMETERS = {'apikey': OMDb_API_KEY,
              't': movie
             }
    response = requests.get(url = ENDPOINT, params = PARAMETERS)  
    #print('Response Code:', response, '\n')
    new_api_call = response.json()
    #print ('Data Type:', type(api_call), '\n')
    #print(api_call)
    #print()
    movies.append(new_api_call)
```

```python
marvel_movies = []
for i in range(len(movies)):
     marvel_movie ={}
     marvel_movie['Title'] = movies[i]['Title']
     marvel_movie['Year'] = int(movies[i]['Year'])
     marvel_movie['Box Office'] = int(movies[i]['BoxOffice'].strip('$').replace(',', ''))
     try:
         marvel_movie['Rotten Tomatoes'] = int(movies[i]['Ratings'][1]['Value'].strip('%'))
     except:
         marvel_movie['Rotten Tomatoes'] = 86 # The Fantastic Four: First Steps Rotten Tomatoes Score missing from API
     marvel_movie['Franchise'] = 'Marvel'
     #print(marvel_movie['Rotten Tomatoes'])
     #print(type((marvel_movie['Rotten Tomatoes])
     marvel_movies.append(marvel_movie)
#print(marvel_movies)
```

```python
marvel_movies_display = pd.DataFrame(marvel_movies)
marvel_movies_display
```

```python
marvel_movies_display.to_csv('marvel_movies.csv')
```
